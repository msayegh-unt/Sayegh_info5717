```python
# Question 1.1
"""
Calculate the number of words that start with a vowel (a, e, i, o, and u). Print
out a list of these words, along with their frequency (number of times they appear in
the text). The list should be in decreasing order of frequency.
"""
# Given Text
Text = "Harry lay in his dark cupboard much later, wishing he had a watch. He didn't know what time it was and he couldn't be sure the Dursleys were asleep yet. Until they were, he couldn't risk sneaking to the kitchen for some food. He'd lived with the Dursleys almost ten years, ten miserable years, as long as he could remember, ever since he'd been a baby and his parents had died in that car crash. He couldn't remember being in the car when his parents had died. Sometimes, when he strained his memory during long hours in his cupboard, he came up with a strange vision: a blinding flash of green light and a burning pain on his forehead.  This, he supposed, was the crash, though he couldn't imagine where all the green light came from. He couldn't remember his parents at all. His aunt and uncle never spoke about them, and of course he was forbidden to ask questions. There were no photographs of them in the house. "

def remove_noise(x):
    result = []
    bad_chars = [';',"''",':','.','<','>','/','?','!','@','#','$',
             '%','^','&','*','(',')','-','_','+','=',' ',',']
    test = x.split()
    for y in test:
        bad_temp = ''
        for i in bad_chars:
            if (i in y):
                y = y.replace(i,'').lower()
        result.append(y)
    return result

text_strip = remove_noise(Text)
dict = {}
for x in text_strip:
    if x[0] in ['a','e','i','o','u']:
        if dict.get(x) == None:
            dict[x] = 1
        else:
            y = int(dict.get(x)) +1
            dict.pop(x,y)
            dict[x] = y

print(sorted(dict.items(), key = 
             lambda kv:(kv[1], kv[0]), reverse=True))
        
```

```python
"""
Question 2. (40 points). I am interested in knowing how the climate changes in terms of
temperature and precipitation. The U.S. climate data site contains climate data for Denton,
Texas since 2009. I would like you to do some calculation and comparison between the data
in 2010 and 2017 in order to answer the following two research questions:

RQ1: Is 2017 significantly different from 2010 on temperature and precipitation in the months
January-June?

RQ2: Is June 2017 significantly hotter than June 2010?

(1) (6 points). Create two files based on data published at U.S. climate data
(http://www.usclimatedata.com/climate/denton/texas/united-states/ustx0353):

➢ File A (should be called 2010-Jan-June.txt, or 2010-Jan-June.csv) contains daily
weather data from January 1, 2010 to June 30, 2010;

➢ File B (called 2017-Jan-June.txt, or 2017-Jan-June.csv) contains daily weather data
from January 1, 2017 to June 30, 2017;

To find the data, go to “History” tab of the above page, select the right year and month. You
will see the data being presented to you.

The final format of each result file should look like the following:
<Date-month>, <High>, <Low>, <Precip>
1-Jan,55,33,0.08

2-Jan,55,33,0.12
……
1-June,80,56,0,15
…
The delimiter can be comma (,) or whitespace. Make sure you round the numbers for the
temperature so there is no decimal points.

(2) (12 points). Write a program to calculate the mean, median, and standard deviation of
high temperature, low temperature and precipitation of each file, and output the results
in the following format:

File name mean median standard deviation
2010-Jan-June.txt ----- ----- -----
2017-Jan-June.txt

(3) (16 points). In order to answer the first research question, we would like to conduct
some statistical tests. Take File A and File B, conduct a T-test on TWO RELATED
samples (you can use scipy.stats.ttest_rel: http://docs.scipy.org/doc/scipy0.14.0/reference/generated/scipy.stats.ttest_rel.html) on Temperature High,
Temperature Low, and Precipitation to find out whether there is a significant difference
between these scores. Report your results in statements after the program using the
docstring.

(4) (6 points). Describe how you can answer research question 2 and what your answer is

"""
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import csv

# State variables
n_rows = 0
n_columns = 0
column_names = []

# Set years for file creation
year = [2010,2017]

# Iterate across years
for y in year:
    # Reset the result list
    result = []
    # Using 6 due to fixed number of months in problem
    for counter in range(6):
        
        # Poll the site for html
        url = 'https://www.usclimatedata.com/climate/denton/texas/united-states/ustx0353/' + str(y) + '/' + str(counter + 1)
        response = requests.get(url)
        
        soup = BeautifulSoup(response.text, 'lxml') # Parse the HTML as a string

        table = soup.find_all(class_="daily_climate_table")[1] # Grab the first table

        # Find number of rows and columns
        # we also find the column titles if we can
        for row in table.find_all('tr'):

            # Determine the number of rows in the table
            td_tags = row.find_all('td')
            if len(td_tags) > 0:
                n_rows+=1
                if n_columns == 0:
                    # Set the number of columns for our table
                    n_columns = len(td_tags)

            # Handle column names if we find them
            th_tags = row.find_all('th') 
            if len(th_tags) > 0 and len(column_names) == 0:
                for th in th_tags:
                    column_names.append(th.get_text())

        columns = column_names if len(column_names) > 0 else range(0,n_columns)    
        new_table = pd.DataFrame(columns = columns, index = range(0,n_rows))

        #DataFrame(columns = columns,index= range(0,n_rows))
        row_marker = -1
        for row in table.find_all('tr'):
            column_marker = 0
            columns = row.find_all('td')

            for column in columns:
                new_table.iat[row_marker,column_marker] = column.get_text()
                column_marker += 1
            row_marker += 1
            if row_marker == n_rows:
                break
        result.append(new_table.to_csv(index=False))
   
    # Sets some variables
    result_clean1 = ''
    result_clean2 = []
    for a in result:
        counter = 0
        # Cleans result for end of line and missing commas
        for b in a:
            if a[counter:counter+1] == '\n':
                result_clean1 += ','
            elif a[counter-1:counter] == '\n':
                continue
            else: result_clean1 += b

    # Create temporary file from table
    result_file = 'tempfile' + str(y) + '.txt'

    FileName = open(result_file, "w+")
    with open(result_file, "w+") as FileName:
        for line in result_clean1:
            FileName.writelines(line)

    # Read temporary file to clean lines
    result_line = []
    line_count = 0
    with open(result_file) as File:
        for line in File:
            if line_count == 0:
                result_line.append(line[0:34])
                line_count += 1
            elif line == ',,,,,\n':
                line_count += 1
            elif line == 'Day,High(°F),Low(°F),Precip.(inch),Snow(inch),Snowdepth(inch)\n':
                line_count += 1
            else: result_line.append(line)

    # Remove extra data in data lines
    result_line2 = []
    line_count = 0
    for c in result_line:
        if line_count == 0:
            header = c.split(',')
            counter = 0
            for f in header:
                if counter < 3:
                    result_line2.append(f)
                    counter += 1
                elif counter == 3:
                    result_line2.append(f)
            line_count += 1
        else:
            d = c.split(",")
            string_count = 0
            for e in d:
                if string_count == 0:
                    result_line2.append(e)
                    string_count += 1
                elif string_count < 3:
                    result_line2.append(int(round(float(e),0)))
                    string_count += 1
                elif string_count == 3: 
                    result_line2.append(e)
                    string_count += 1
            line_count += 1

# Split list into lines for file
    composite_list = [result_line2[x:x+4] for x in range(0, len(result_line2),4)]

    # Write to .csv file
    result_file = 'File' + str(y) + '.csv'
    with open(result_file, "w+") as FileName:
        writer = csv.writer(FileName, delimiter=',')
        for line in composite_list:
            writer.writerow(line)
        print(str(y) + " File Created")
```

```python
import statistics

# Establish Variables
temp_avg = 0.0
temp_median = 0.0
temp_stdev = 0.0

def mean_calculation(file_year,column):
    list_temp = []
    counter = 0
    test_file = 'File' + str(file_year) + '.csv'
    with open(test_file) as File:
        for line in File:
            if counter == 0:
                counter += 1
            else: 
                list_strip = line.rstrip()
                list_temp.append(list_strip.split(',')[column])
        return list_temp

def convert_to_int(list_str):
    list_str = [float(i) for i in list_str] 
    return list_str

def get_stats(list_int):
    t_avg = round(statistics.mean(list_int),2)
    t_median = round(statistics.median(list_int),2)
    t_stdev = round(statistics.stdev(list_int),2)
    return(t_avg,t_median,t_stdev)

# Run Calculations
calc_type = ['High Temp','Low Temp','Precipitation']
for y in range(len(calc_type)):
    year = [2010,2017]

    # Print headers
    print()
    print(calc_type[y])
    print()
    print('File Name' + ' '*10 + 'Mean' + ' '*5 + 'Median' + ' '*5 + 'Standard Deviation')
    print()
    for x in year:
        stats = []
        list_str = mean_calculation(x,y+1)
        list_int = convert_to_int(list_str) 
        stats = get_stats(list_int)

        # Print reuslts
        print('File' + str(x) + '.csv' + ' '*7 + str(stats[0]) + ' '*4 + str(stats[1]) + ' '*10 + str(stats[2]))
     
```

```python
"""

Question 3. (30 points). We have discussed how we could create a Python class, and how to
define the properties, functions, and data types it would require. For this question, you 
are required to create and implement the class Student in the 
“<LastName>_assognment-threecode.ipynb” file. Your class should be able to do the 
following: 

(1) (6 pts) Every instance of the Student class should have the following attributes:
➢ First Name
➢ Last Name
➢ Middle Name
➢ euid, by default euid is blank.
➢ GPA, by default the GPA is set to 0
➢ Classes taken. This object should store the class number (such as INFO 5717), the
semester it was taken it (such as "Fall 2019"), and the grade the student obtained.
HINT: use a collection data type.

(2) (6 pts) Write a program in the “<LastName>_assognment-two-code.ipynb” file that
asks the user to enter information for three fictional students. The program should
prompt the user to ask for the student's first, middle, and last names. This information
should be stored in an object of type Student.

(3) (6 pts) Write a function for the Student class that assigns each student a euid. The 
euid should be composed of three letters (the first three letters of the student's first, 
middle, and last names) and then 4 digits chosen randomly. For example, if a student's name is
Jennifer Lynn Meyers her euid could be "jlm8940". Use this function to print out the
names and euids of the students you created in (2).

(4) (6 pts) Write a function for the Student class called "register". This function asks the
user to enter a class number, a semester, and a grade for each student. Then this
information is added to the Student object. Call this function for the students you
created in part (2). Each student should be registered for at least two courses but can
have more than.

(5) (6 pts) Write a function for the Student class that calculates a student's GPA and prints
it out. Use it to print out the names and GPAs of the student's you created in (2). 
"""
```

```python
from random import randint

# Create class for Student
class Student: 

    # init method or constructor  
    def __init__(self, FirstName, LastName, MiddleName): 
        self.FirstName = FirstName
        self.LastName = LastName
        self.MiddleName = MiddleName
        self.Euid = FirstName[0] + MiddleName[0] + LastName[0] + format(randint(0,9999),'03')
        self.Registration = []
  
    # Methods
    def FullName(self): 
        return self.FirstName + ' ' + self.MiddleName + ' ' + self.LastName 
    
    def Register(self):
        print('This allows you to register for a class. Please enter the following: \n')
        class_list = []
        class_num = input('Class Number: \n')
        class_list.append(class_num)
        semester = input('Semester: \n')
        class_list.append(semester)
        grade_try = True
        while grade_try == True:
            try:
                grade = float(input('Grade: \n'))
                grade_try = False
            except ValueError:
                print('Not an appropriate grade number')
        class_list.append(float(grade))
        self.Registration.append(class_list)
        
    def gpa(self):
        calc = 0.0
        res = 0.0
        if self.Registration == []:
            return calc
        for t in self.Registration:
            calc += t[2]
        res = calc/len(self.Registration)
        return res

# Input names for students function
def get_info():
    first_name = input('Enter First Name: \n')
    last_name = input('Enter Last Name: \n')
    middle_name = input('Enter Middle Name: \n')
    return (first_name, last_name, middle_name)

# Predicated on running for 3 students
students = ['Student1','Student2','Student3']

# Input names and create objects
for x in range(len(students)):
    name = get_info()
    students[x] = Student(name[0],name[1],name[2])

for x in range(len(students)):
    counter = 0
    status = True
    while counter < 2:
        while status == True:
            try:
                print(students[x].FullName())
                counter = int(input('Number of classes to enter (must be atleast 2):\n'))
                if counter >= 2:
                    status = False
            except ValueError:
                print('Not an appropriate number')
    for y in range(counter):
        students[x].Register()

for x in range(len(students)):
    print(students[x].FullName() + ' (EUID: ' + students[x].Euid + ')')
    print(students[x].Registration)
    gpa = students[x].gpa()
    print('GPA: ' + str(round(gpa,2)))
```
